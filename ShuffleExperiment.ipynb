```python
import random, math


def shuffle_sort(n):
    number_list = [i for i in range(1, n + 1)]
    random_numbers = random.sample(range(0, n**3), n)
    zipped_list = dict(zip(number_list, random_numbers))
    sortlist = sorted(zipped_list, key=lambda i: zipped_list[i])
    return sortlist


def shuffle_swap(n):
    number_list = [i for i in range(1, n + 1)]
    for i in range(len(number_list)):
        random_number = random.randint(i, n - 1)
        number_list[i], number_list[random_number] = (
            number_list[random_number],
            number_list[i],
        )
    return number_list


def distibution(n):
    final_set = set()
    for i in range(100):
        new = shuffle_swap(n)
        final_set.add(tuple(new))
    return len(final_set)


n = 5
factorial = math.factorial(n)
print("Factorial: {}".format(factorial))

# Check distribution for shuffle_sort
print("Distribution (shuffle_swap): {}".format(distibution(n)))
```

```python
import random
import matplotlib.pyplot as plt

def shuffle_swap(n):
    number_list = [i for i in range(1, n + 1)]
    for i in range(len(number_list)):
        random_number = random.randint(i, n - 1)
        number_list[i], number_list[random_number] = (
            number_list[random_number],
            number_list[i],
        )
    return number_list

def distribution(n, num_samples):
    counts = {}
    for i in range(num_samples):
        new = shuffle_swap(n)
        new_tuple = tuple(new)
        counts[new_tuple] = counts.get(new_tuple, 0) + 1
    return counts

n = 5
num_samples = 1000

counts = distribution(n, num_samples)

plt.bar(range(len(counts)), counts.values())
plt.xlabel("Unique Shuffled Lists")
plt.ylabel("Frequency")
plt.title("Distribution of Shuffled Lists")
plt.xticks(range(len(counts)), counts.keys(), rotation=90)
plt.show()
```
